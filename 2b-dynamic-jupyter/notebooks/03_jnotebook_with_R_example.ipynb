## Develop a new report
We'll use the code chunk below to develop a new report. It should contain a title, two text blocks and two code blocks. The code requires the dplyr and babynames packages. Uncomment the `install.packages` lines if you need to install them.

```r
#install.packages('dplyr')
#install.packages('babynames')
#install.packages('ggplot2')

library(dplyr)
library(babynames)
library(ggplot2)
```

# Plot name frequency over time 

```r
babynames %>%
  filter(name == "Mary" & sex == "F") %>%
  ggplot(aes(x = year, y = n)) +
  geom_line() + labs(x = "Year", y = "Number Born", 
                     title = "Mary as a Girl-Name")
```

# Compare name use by gender over time

```r
babynames %>%
  filter(name == "Leslie") %>%
  ggplot(aes(x = year, y = n)) +
  geom_line(aes(color = sex)) + labs(x = "Year", y = "Number Born", 
                                     title = "Leslie, by Sex")
                                     
```
